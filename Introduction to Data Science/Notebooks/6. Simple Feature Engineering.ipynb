# Simple Feature Engineering

Pada notebook ini, kita akan membuat fitur-fitur baru dari data Titanic yang sudah ada. Dataset `titanic_cleaned.csv` akan diperkaya dan disimpan sebagai `titanic_engineered.csv`.

## 1. Memuat Dataset

```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
print(f"Dimensi awal: {df.shape}")
print(f"Kolom: {df.columns.tolist()}")
```

## 2. Family Size & Is Alone

```python
df["family_size"] = df["sibsp"] + df["parch"] + 1
df["is_alone"] = (df["family_size"] == 1).astype(int)

print("Family size vs Survival:")
print(df.groupby("family_size")["survived"].mean().round(4))
print(f"\nIs alone vs Survival:")
print(df.groupby("is_alone")["survived"].mean().round(4))
```

## 3. Age Group

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Anak", "Remaja", "Dewasa_Muda", "Dewasa", "Lansia"]
)
print(df["age_group"].value_counts().sort_index())
print(f"\nSurvival per age group:")
print(df.groupby("age_group")["survived"].mean().round(4))
```

## 4. Fare Category

```python
df["fare_category"] = pd.cut(
    df["fare"],
    bins=[-1, 7.91, 14.454, 31.0, 600],
    labels=["Murah", "Menengah", "Mahal", "Sangat_Mahal"]
)
print(df.groupby("fare_category")["survived"].mean().round(4))
```

## 5. Mengekstrak Title dari Nama

```python
df["title"] = df["name"].str.extract(r",\s*([^\.]+)\.")

print("Semua title:")
print(df["title"].value_counts())
```

```python
# Kelompokkan title langka
def kelompokkan_title(title):
    if title in ["Mr"]:
        return "Mr"
    elif title in ["Miss", "Mlle", "Ms"]:
        return "Miss"
    elif title in ["Mrs", "Mme"]:
        return "Mrs"
    elif title in ["Master"]:
        return "Master"
    else:
        return "Lainnya"

df["title"] = df["title"].apply(kelompokkan_title)
print("Title setelah pengelompokan:")
print(df["title"].value_counts())
print(f"\nSurvival per title:")
print(df.groupby("title")["survived"].mean().round(4).sort_values(ascending=False))
```

## 6. Menghapus Kolom yang Tidak Diperlukan

```python
df = df.drop(columns=["passenger_id", "name"])
print(f"Kolom akhir: {df.columns.tolist()}")
```

## 7. Menyimpan Dataset

```python
df.to_csv("titanic_engineered.csv", index=False)
print("Dataset disimpan sebagai titanic_engineered.csv")
print(f"Dimensi: {df.shape}")
print(f"\nPerubahan: 10 kolom -> {df.shape[1]} kolom")
print("Kolom baru: family_size, is_alone, age_group, fare_category, title")
print("Kolom dihapus: passenger_id, name")
df.head()
```
