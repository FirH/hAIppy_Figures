# Understanding Data & Dataset

Pada notebook ini, kita akan memuat dataset Titanic, menjelajahi strukturnya, dan menyimpannya sebagai `titanic_raw.csv`.

Dataset Titanic berisi informasi 891 penumpang kapal RMS Titanic, termasuk status keselamatan, kelas tiket, umur, dan atribut lainnya.

## 1. Import Library

```python
import pandas as pd
import numpy as np
print("Library berhasil diimpor!")
```

## 2. Memuat Dataset

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(f"Dataset berhasil dimuat!")
print(f"Dimensi: {df.shape[0]} baris, {df.shape[1]} kolom")
```

## 3. Melihat Data Awal

```python
df.head()
```

## 4. Informasi Dataset

```python
df.info()
```

## 5. Deskripsi Kolom

| Kolom | Deskripsi |
|---|---|
| PassengerId | ID unik penumpang |
| Survived | Status keselamatan (0=Tidak, 1=Ya) |
| Pclass | Kelas tiket (1=Pertama, 2=Kedua, 3=Ketiga) |
| Name | Nama penumpang |
| Sex | Jenis kelamin |
| Age | Umur penumpang |
| SibSp | Jumlah saudara/pasangan di kapal |
| Parch | Jumlah orangtua/anak di kapal |
| Ticket | Nomor tiket |
| Fare | Harga tiket |
| Cabin | Nomor kabin |
| Embarked | Pelabuhan (C=Cherbourg, Q=Queenstown, S=Southampton) |

## 6. Mendeteksi Missing Values

```python
print("Missing values per kolom:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

## 9. Filtering Data

```python
# Penumpang yang selamat
selamat = df[df["Survived"] == 1]
print(f"Jumlah penumpang selamat: {len(selamat)}")
selamat[["Name", "Sex", "Age", "Pclass"]].head()
```

```python
# Penumpang wanita kelas 1 yang selamat
filter_hasil = df[(df["Sex"] == "female") & (df["Pclass"] == 1) & (df["Survived"] == 1)]
print(f"Jumlah: {len(filter_hasil)}")
filter_hasil[["Name", "Age", "Fare"]].head()
```

## 10. Menyimpan Dataset

```python
df.to_csv("titanic_raw.csv", index=False)
print("Dataset disimpan sebagai titanic_raw.csv")
print(f"Kolom: {df.columns.tolist()}")
print(f"Dimensi: {df.shape}")
```
