# Descriptive Statistics

Pada notebook ini, kita akan menganalisis karakteristik dataset Titanic menggunakan ukuran-ukuran statistik deskriptif. Dataset yang digunakan adalah `titanic_cleaned.csv`.

## 1. Memuat Dataset

```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
print(f"Dimensi: {df.shape}")
df.head()
```

## 2. Ringkasan Statistik

```python
df.describe().round(2)
```

## 3. Ukuran Pemusatan (Central Tendency)

```python
# Mean
print("=== MEAN ===")
print(f"Rata-rata umur: {df['age'].mean():.2f}")
print(f"Rata-rata fare: {df['fare'].mean():.2f}")
print(f"Rata-rata survived: {df['survived'].mean():.4f} ({df['survived'].mean()*100:.1f}% selamat)")
```

```python
# Median
print("=== MEDIAN ===")
print(f"Median umur: {df['age'].median()}")
print(f"Median fare: {df['fare'].median()}")
print(f"\nPerhatikan: mean fare ({df['fare'].mean():.2f}) >> median fare ({df['fare'].median():.2f})")
print("-> Distribusi fare sangat miring ke kanan (positively skewed)")
```

```python
# Mode
print("=== MODE ===")
print(f"Modus umur: {df['age'].mode()[0]}")
print(f"Modus sex: {df['sex'].mode()[0]}")
print(f"Modus pclass: {df['pclass'].mode()[0]}")
print(f"Modus embarked: {df['embarked'].mode()[0]}")
```

## 4. Ukuran Penyebaran (Dispersion)

```python
# Standard Deviation & Range
print("=== STANDARD DEVIATION ===")
print(f"Std umur: {df['age'].std():.2f}")
print(f"Std fare: {df['fare'].std():.2f}")

print(f"\n=== RANGE ===")
print(f"Range umur: {df['age'].min():.1f} - {df['age'].max():.1f} = {df['age'].max() - df['age'].min():.1f}")
print(f"Range fare: {df['fare'].min():.2f} - {df['fare'].max():.2f} = {df['fare'].max() - df['fare'].min():.2f}")
```

## 5. Quartile, IQR, dan Deteksi Outlier

```python
q1 = df["fare"].quantile(0.25)
q3 = df["fare"].quantile(0.75)
iqr = q3 - q1
batas_bawah = q1 - 1.5 * iqr
batas_atas = q3 + 1.5 * iqr

print(f"Q1 fare: {q1:.2f}")
print(f"Q3 fare: {q3:.2f}")
print(f"IQR: {iqr:.2f}")
print(f"Batas outlier: {batas_bawah:.2f} - {batas_atas:.2f}")

outlier = df[(df["fare"] < batas_bawah) | (df["fare"] > batas_atas)]
print(f"Jumlah outlier fare: {len(outlier)}")
```

## 6. Statistik Data Kategorikal

```python
print("=== Distribusi Sex ===")
print(df["sex"].value_counts())
print(f"\n=== Distribusi Pclass ===")
print(df["pclass"].value_counts().sort_index())
print(f"\n=== Distribusi Embarked (%) ===")
print((df["embarked"].value_counts(normalize=True) * 100).round(1))
```

## 7. Perbandingan Mean vs Median

```python
print("=== Perbandingan Mean vs Median ===")
for kolom in ["age", "fare", "sibsp", "parch"]:
    mean_val = df[kolom].mean()
    median_val = df[kolom].median()
    arah = "kanan" if mean_val > median_val else "kiri" if mean_val < median_val else "simetris"
    print(f"{kolom}: mean={mean_val:.2f}, median={median_val:.2f} -> miring ke {arah}")
```

## 8. Simpan Dataset

Dataset tidak berubah struktural pada tahap ini.

```python
df.to_csv("titanic_cleaned.csv", index=False)
print(f"Dataset: titanic_cleaned.csv ({df.shape[0]} baris, {df.shape[1]} kolom)")
```
