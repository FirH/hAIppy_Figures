# Data Aggregation & Grouping

Pada notebook ini, kita akan merangkum data Titanic berdasarkan kategori menggunakan `groupby()` dan `pivot_table()`.

## 1. Memuat Dataset

```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
print(f"Dimensi: {df.shape}")
```

## 2. GroupBy Dasar - Tingkat Keselamatan per Kelas

```python
survival_rate = df.groupby("pclass")["survived"].mean()
print(survival_rate.round(4))
```

## 3. Berbagai Fungsi Agregasi

```python
# Jumlah penumpang per kelas
print("Jumlah penumpang per kelas:")
print(df.groupby("pclass")["passenger_id"].count())
```

```python
# Rata-rata fare dan umur per kelas
print("Rata-rata fare per kelas:")
print(df.groupby("pclass")["fare"].mean().round(2))
print("\nRata-rata umur per sex:")
print(df.groupby("sex")["age"].mean().round(2))
```

## 4. Named Aggregation

```python
hasil = df.groupby("pclass").agg(
    jumlah_penumpang=("passenger_id", "count"),
    tingkat_selamat=("survived", "mean"),
    rata_umur=("age", "mean"),
    rata_fare=("fare", "mean")
).round(2)
hasil
```

## 5. Grouping dengan Beberapa Kategori

```python
hasil = df.groupby(["pclass", "sex"])["survived"].mean().round(4)
print(hasil)
```

## 6. Sorting Hasil Aggregasi

```python
fare_per_embarked = df.groupby("embarked")["fare"].mean().sort_values(ascending=False)
print(fare_per_embarked.round(2))
```

## 7. Pivot Table - Keselamatan per Kelas dan Sex

```python
pivot = df.pivot_table(
    values="survived",
    index="pclass",
    columns="sex",
    aggfunc="mean"
).round(4)
pivot
```

```python
# Pivot table - Fare per Embarked dan Pclass
pivot_fare = df.pivot_table(
    values="fare",
    index="embarked",
    columns="pclass",
    aggfunc="mean",
    fill_value=0
).round(2)
pivot_fare
```

## 8. Filtering Hasil Aggregasi

```python
survival = df.groupby(["pclass", "sex"])["survived"].mean()
print("Kelompok dengan survival rate > 50%:")
print(survival[survival > 0.5].round(4))
```

## 9. Simpan Dataset

```python
df.to_csv("titanic_cleaned.csv", index=False)
print(f"Dataset: titanic_cleaned.csv ({df.shape[0]} baris, {df.shape[1]} kolom)")
```
