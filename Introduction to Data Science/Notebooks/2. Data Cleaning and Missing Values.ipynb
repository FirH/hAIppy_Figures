# Data Cleaning & Missing Values

Pada notebook ini, kita akan membersihkan `titanic_raw.csv` dengan menangani missing values, menghapus kolom yang tidak diperlukan, dan menyimpan hasilnya sebagai `titanic_cleaned.csv`.

## 1. Memuat Dataset

```python
import pandas as pd

df = pd.read_csv("titanic_raw.csv")
print(f"Dimensi awal: {df.shape}")
df.head()
```

## 2. Mendeteksi Missing Values

```python
print("Missing values per kolom:")
print(df.isnull().sum())
```

```python
# Persentase missing values
persen_missing = df.isnull().sum() / len(df) * 100
print("\nPersentase missing values (kolom yang memiliki missing):")
print(persen_missing[persen_missing > 0].round(2))
```

## 3. Menghapus Kolom Cabin (77% missing)

```python
df = df.drop(columns=["Cabin"])
print(f"Kolom setelah drop Cabin: {df.columns.tolist()}")
```

## 4. Mengisi Missing Values - Age (median)

```python
median_age = df["Age"].median()
print(f"Median Age: {median_age}")

df["Age"] = df["Age"].fillna(median_age)
print(f"Missing Age setelah fillna: {df['Age'].isnull().sum()}")
```

## 5. Mengisi Missing Values - Embarked (modus)

```python
print("Distribusi Embarked:")
print(df["Embarked"].value_counts())

modus_embarked = df["Embarked"].mode()[0]
print(f"\nModus Embarked: {modus_embarked}")

df["Embarked"] = df["Embarked"].fillna(modus_embarked)
print(f"Missing Embarked setelah fillna: {df['Embarked'].isnull().sum()}")
```

## 6. Verifikasi - Semua Missing Values Tertangani

```python
print("Missing values setelah cleaning:")
print(df.isnull().sum())
```

## 7. Memeriksa Duplikat

```python
jumlah_duplikat = df.duplicated().sum()
print(f"Jumlah baris duplikat: {jumlah_duplikat}")
df = df.drop_duplicates()
```

## 8. Menghapus Kolom Ticket (tidak informatif)

```python
df = df.drop(columns=["Ticket"])
print(f"Kolom setelah drop Ticket: {df.columns.tolist()}")
```

## 9. Rename Kolom ke snake_case

```python
df = df.rename(columns={
    "PassengerId": "passenger_id",
    "Survived": "survived",
    "Pclass": "pclass",
    "Name": "name",
    "Sex": "sex",
    "Age": "age",
    "SibSp": "sibsp",
    "Parch": "parch",
    "Fare": "fare",
    "Embarked": "embarked"
})
print(f"Kolom baru: {df.columns.tolist()}")
df.head()
```

## 10. Menyimpan Dataset yang Sudah Dibersihkan

```python
df.to_csv("titanic_cleaned.csv", index=False)
print("Dataset disimpan sebagai titanic_cleaned.csv")
print(f"Dimensi akhir: {df.shape}")
print(f"\nPerubahan: 12 kolom -> {df.shape[1]} kolom (hapus Cabin, Ticket; isi Age, Embarked; rename)")
```
