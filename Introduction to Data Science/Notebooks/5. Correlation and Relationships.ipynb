# Correlation & Relationships

Pada notebook ini, kita akan mengukur dan memahami hubungan antar variabel dalam dataset Titanic.

## 1. Memuat Dataset

```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")
print(f"Dimensi: {df.shape}")
print(f"Kolom numerik: {df.select_dtypes(include='number').columns.tolist()}")
```

## 2. Korelasi Antara Dua Kolom

```python
print(f"Korelasi fare vs survived: {df['fare'].corr(df['survived']):.4f}")
print(f"Korelasi pclass vs survived: {df['pclass'].corr(df['survived']):.4f}")
print(f"Korelasi age vs survived: {df['age'].corr(df['survived']):.4f}")
print(f"Korelasi pclass vs fare: {df['pclass'].corr(df['fare']):.4f}")
```

## 3. Matriks Korelasi

```python
kolom_analisis = ["survived", "pclass", "age", "sibsp", "parch", "fare"]
matriks = df[kolom_analisis].corr().round(3)
matriks
```

## 4. Korelasi Terhadap Variabel Target (survived)

```python
korelasi_survived = df[kolom_analisis].corr()["survived"].drop("survived").sort_values(ascending=False)
print(korelasi_survived.round(4))
```

## 5. Cross Tabulation (Variabel Kategorikal)

```python
# Sex vs Survived
tabel = pd.crosstab(df["sex"], df["survived"])
tabel.columns = ["Tidak Selamat", "Selamat"]
print(tabel)
```

```python
# Persentase per baris
tabel_persen = pd.crosstab(df["sex"], df["survived"], normalize="index") * 100
tabel_persen.columns = ["Tidak Selamat (%)", "Selamat (%)"]
print(tabel_persen.round(1))
```

```python
# Embarked vs Survived
tabel_emb = pd.crosstab(df["embarked"], df["survived"], normalize="index") * 100
tabel_emb.columns = ["Tidak Selamat (%)", "Selamat (%)"]
print(tabel_emb.round(1))
```

## 6. Korelasi per Kelompok

```python
for kelas in sorted(df["pclass"].unique()):
    subset = df[df["pclass"] == kelas]
    kor = subset["age"].corr(subset["survived"])
    print(f"Korelasi age vs survived (Kelas {kelas}): {kor:.4f}")

print()
for gender in ["male", "female"]:
    subset = df[df["sex"] == gender]
    kor = subset["fare"].corr(subset["survived"])
    print(f"Korelasi fare vs survived ({gender}): {kor:.4f}")
```

## 7. Simpan Dataset

```python
df.to_csv("titanic_cleaned.csv", index=False)
print(f"Dataset: titanic_cleaned.csv ({df.shape[0]} baris, {df.shape[1]} kolom)")
```
